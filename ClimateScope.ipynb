```python
df['last_updated'] = pd.to_datetime(df['last_updated'])
df['month'] = df['last_updated'].dt.month
```

```python
import matplotlib.pyplot as plt

monthly_avg = df.groupby('month')['temperature_celsius'].mean()

plt.plot(monthly_avg)
plt.xlabel("Month")
plt.ylabel("Average Temperature")
plt.title("Monthly Average Temperature")
plt.show()
```

```python
df[['last_updated','month']].head()
```

```python
monthly_humidity = df.groupby('month')['humidity'].mean()

plt.figure()
plt.plot(monthly_humidity)
plt.xlabel("Month")
plt.ylabel("Average Humidity")
plt.title("Monthly Average Humidity")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(df['temperature_celsius'], df['humidity'])
plt.xlabel("Temperature (Celsius)")
plt.ylabel("Humidity")
plt.title("Temperature vs Humidity")
plt.show()
```
